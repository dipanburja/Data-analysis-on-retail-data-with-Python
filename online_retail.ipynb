Step 3 – Data Cleaning

```python
# 1. Drop rows where CustomerID is null (we can't track behavior without IDs)
df_clean = df.dropna(subset=['CustomerID'])

# 2. Remove cancelled orders (InvoiceNo starting with 'C')
# We use .astype(str) to ensure we can perform string operations
df_clean = df_clean[~df_clean['InvoiceNo'].astype(str).str.startswith('C')]

# 3. Filter out non-positive Quantities and UnitPrices
df_clean = df_clean[(df_clean['Quantity'] > 0) & (df_clean['UnitPrice'] > 0)]

# 4. Create TotalPrice column
df_clean['TotalPrice'] = df_clean['Quantity'] * df_clean['UnitPrice']

# 5. Extract Time features
df_clean['Month'] = df_clean['InvoiceDate'].dt.month_name()
df_clean['DayOfWeek'] = df_clean['InvoiceDate'].dt.day_name()
df_clean['Hour'] = df_clean['InvoiceDate'].dt.hour

# Display the new shape and first few rows
print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_clean.shape}")
df_clean.head()
```

Descriptive Statistics

```python
# Focus on the numeric columns that drive revenue
stats_cols = ['Quantity', 'UnitPrice', 'TotalPrice']
descriptive_stats = df_clean[stats_cols].describe().T

# Add Skewness to see how "long" the tail of your data is
descriptive_stats['skew'] = df_clean[stats_cols].skew()
print(descriptive_stats)
```

Identifying Outliers (The IQR Method)
LowerBound=Q1−1.5×IQR
UpperBound=Q3+1.5×IQR

```python
# Example for TotalPrice outliers
Q1 = df_clean['TotalPrice'].quantile(0.25)
Q3 = df_clean['TotalPrice'].quantile(0.75)
IQR = Q3 - Q1

outliers = df_clean[(df_clean['TotalPrice'] < (Q1 - 1.5 * IQR)) |
                    (df_clean['TotalPrice'] > (Q3 + 1.5 * IQR))]

print(f"Number of outlier transactions: {len(outliers)}")
```

Visualizing Trends (The Zoom-In Strategy)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure months are in chronological order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

monthly_sales = df_clean.groupby('Month')['TotalPrice'].sum().reindex(month_order)

plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette='viridis')
plt.title('Monthly Revenue (2010-2011)', fontsize=15)
plt.xticks(rotation=45)
plt.ylabel('Total Revenue (£)')
plt.show()
```

```python
hourly_orders = df_clean.groupby('Hour')['InvoiceNo'].nunique()

plt.figure(figsize=(10, 5))
sns.lineplot(x=hourly_orders.index, y=hourly_orders.values, marker='o', color='tab:red')
plt.title('Busiest Shopping Hours (Number of Invoices)', fontsize=14)
plt.xticks(range(6, 21)) # The dataset usually spans 6am to 8pm
plt.grid(True, alpha=0.3)
plt.show()
```

6. Top Sellers & Countries (Excl. UK)

```python
# Top 10 Countries by Revenue (Excluding UK)
top_countries = df_clean[df_clean['Country'] != 'United Kingdom'] \
                .groupby('Country')['TotalPrice'].sum() \
                .sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_countries.plot(kind='barh', color='skyblue')
plt.title('Top 10 International Markets by Revenue', fontsize=14)
plt.gca().invert_yaxis() # Highest revenue at the top
plt.show()
```

Step 7: Deep Dive into the "Whales"

```python
# Identify the 5 most expensive transactions
top_5_extremes = df_clean.sort_values(by='TotalPrice', ascending=False).head(5)

print("--- Top 5 Extreme Transactions ---")
print(top_5_extremes[['Description', 'Quantity', 'UnitPrice', 'TotalPrice', 'CustomerID', 'Country']])
```
